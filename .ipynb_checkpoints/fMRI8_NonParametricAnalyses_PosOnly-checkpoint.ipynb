```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
import statsmodels.sandbox.stats.multicomp as mc
import sklearn
import seaborn as sns
plt.style.use("ggplot")
sns.set_style('whitegrid')
plt.rcParams["font.family"] = "Arial"
import nibabel as nib
import pandas as pd
import sys
import h5py
from importlib import reload

from brainsmash.mapgen.base import Base
from brainsmash.mapgen.stats import pearsonr, pairwise_r
from brainsmash.mapgen.stats import nonparp


glasserfile2 = '../../data/Q1-Q6_RelatedParcellation210.LR.CorticalAreas_dil_Colors.32k_fs_RL.dlabel.nii'
glasser2 = nib.load(glasserfile2).get_data()
glasser2 = np.squeeze(glasser2)
nParcels = 360
```

# Load in data

```python
resultdir = '../../data/results/'

#### Load time scale data
taus_rest = np.loadtxt('../../data/results/Murray_taus_regions_autocorrelation_rest100.txt',delimiter=',')
bad_ind = np.where(taus_rest>50)
taus_rest[bad_ind] = np.nan
# bad_ind = np.where(taus_rest<0)
# taus_rest[bad_ind] = np.nan

#### Load in myelin data
myelinmap = np.squeeze(nib.load('../../data/results/Mean.339.MyelinMap_BC_MSMAll.32k_fs_LR.dscalar.nii').get_data())
nvertices = int(glasser2.shape[0]/2)
glasser_lr = np.hstack((glasser2[nvertices:],glasser2[:nvertices])) # need to flip this original parcellation
myelinmap_parc = np.zeros((nParcels,))
for roi in range(nParcels):
    roi_ind = np.where(glasser_lr==roi+1)[0]
    myelinmap_parc[roi] = np.mean(myelinmap[roi_ind])

#### Load in task activation data
resultdir = '../../data/results/'
h5f = h5py.File(resultdir + 'taskAct_maxActMethod.h5','r')
act_alltasks = h5f['data'][:]
t_act_peak= np.mean(np.abs(stats.ttest_1samp(act_alltasks,0,axis=2)[0]),axis=1)
h5f.close()

#### Load in fc change data
h5f = h5py.File(resultdir + 'taskGBC_maxActMethod.h5','r')
taskgbc = h5f['data'][:]
h5f.close()

h5f = h5py.File(resultdir + 'restGBC_maxActMethod.h5','r')
restgbc = h5f['data'][:]
h5f.close()
fcchange_peak = taskgbc - restgbc

#### Load in Margulies gradient data
# RH
gradient_rh = np.squeeze(nib.load('../../data/results/margulies2016/hcp.tmp.rh.dscalar.nii').get_data())
# Remove all the '0s' in the array
nonzero_ind = np.where(gradient_rh!=0)[0]
gradient_rh = gradient_rh[nonzero_ind].copy()
# LH
gradient_lh = np.squeeze(nib.load('../../data/results/margulies2016/hcp.tmp.lh.dscalar.nii').get_data())
# Remove all the '0s' in the array
nonzero_ind = np.where(gradient_lh!=0)[0]
gradient_lh = gradient_lh[nonzero_ind].copy()
# Glasser atlas is orgnized R->L hemispheres
gradient = np.hstack((gradient_lh,gradient_rh))
# Parcellate
gradient_parcellated = np.zeros((nParcels,))
for roi in range(nParcels):
    roi_ind = np.where(glasser2==roi+1)[0]
    gradient_parcellated[roi] = np.mean(gradient[roi_ind])

#### Load in canonical HRF task activations GLM
h5f = h5py.File(resultdir + 'univariate_act.h5','r')
act_alltasks2 = h5f['data'][:]
t_act = np.mean(np.abs(stats.ttest_1samp(act_alltasks2,0,axis=2)[0]),axis=1)
h5f.close()

#### Load in FIR based task FC
fcchange = np.loadtxt(resultdir + 'conn_taskVrest_gbc_FIR.csv', delimiter=',')
```

___
# Compute surrogate maps using BrainSMASH

```python
geo_dist_rh = np.loadtxt('../../data/BrainSMASH_GeoDesicDistances/RightParcelGeodesicDistmat.txt')
geo_dist_lh = np.loadtxt('../../data/BrainSMASH_GeoDesicDistances/LeftParcelGeodesicDistmat.txt')
n_surrogates = 1000

print('Generate surrogate maps for task activity')
base = Base(t_act[:180],geo_dist_lh)
surrogates_lh = base(n=n_surrogates)
base = Base(t_act[180:],geo_dist_rh)
surrogates_rh = base(n=n_surrogates)
surrogates_tact = np.hstack((surrogates_lh,surrogates_rh))

print('Generate surrogate maps for FC change')
base = Base(np.mean(fcchange,axis=1)[:180],geo_dist_lh)
surrogates_lh = base(n=n_surrogates)
base = Base(np.mean(fcchange,axis=1)[180:],geo_dist_rh)
surrogates_rh = base(n=n_surrogates)
surrogates_fc = np.hstack((surrogates_lh,surrogates_rh))

print('Generate surrogate maps for myelin map')
base = Base(myelinmap_parc[:180],geo_dist_lh)
surrogates_lh = base(n=n_surrogates)
base = Base(myelinmap_parc[180:],geo_dist_rh)
surrogates_rh = base(n=n_surrogates)
surrogates_myelin = np.hstack((surrogates_lh,surrogates_rh))

print('Generate surrogate maps for timescale map')
base = Base(np.nanmean(taus_rest,axis=1)[:180],geo_dist_lh)
surrogates_lh = base(n=n_surrogates)
base = Base(np.nanmean(taus_rest,axis=1)[180:],geo_dist_rh)
surrogates_rh = base(n=n_surrogates)
surrogates_timescale = np.hstack((surrogates_lh,surrogates_rh))

print('Generate surrogate maps for principal gradient')
base = Base(gradient_parcellated[:180],geo_dist_lh)
surrogates_lh = base(n=n_surrogates)
base = Base(gradient_parcellated[180:],geo_dist_rh)
surrogates_rh = base(n=n_surrogates)
surrogates_gradient = np.hstack((surrogates_lh,surrogates_rh))

print('Generate surrogate maps for PEAK task activity')
base = Base(t_act_peak[:180],geo_dist_lh)
surrogates_lh = base(n=n_surrogates)
base = Base(t_act_peak[180:],geo_dist_rh)
surrogates_rh = base(n=n_surrogates)
surrogates_tact_peak = np.hstack((surrogates_lh,surrogates_rh))

print('Generate surrogate maps for PEAK FC change')
base = Base(np.mean(fcchange_peak,axis=1)[:180],geo_dist_lh)
surrogates_lh = base(n=n_surrogates)
base = Base(np.mean(fcchange_peak,axis=1)[180:],geo_dist_rh)
surrogates_rh = base(n=n_surrogates)
surrogates_fc_peak = np.hstack((surrogates_lh,surrogates_rh))
```

# Compute null distribution p-values with correct SA -- regular FC/activity estimates

```python
labels = ['Activity','FC change','Timescale','Myelin','Gradient']
corrmat = np.zeros((5,5))
pmat = np.zeros((5,5))
#### For activity
surrogate_brainmap_corrs = pearsonr(surrogates_tact,np.mean(fcchange,axis=1)).flatten()
test_stat = stats.pearsonr(t_act,np.mean(fcchange,axis=1))[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[0,1], pmat[0,1] = test_stat, pval
#
surrogate_brainmap_corrs = pearsonr(surrogates_tact,np.nanmean(taus_rest,axis=1)).flatten()
test_stat = stats.pearsonr(t_act,np.nanmean(taus_rest,axis=1))[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[0,2], pmat[0,2] = test_stat, pval 
# 
surrogate_brainmap_corrs = pearsonr(surrogates_tact,myelinmap_parc).flatten()
test_stat = stats.pearsonr(t_act,myelinmap_parc)[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[0,3], pmat[0,3] = test_stat, pval 
# 
surrogate_brainmap_corrs = pearsonr(surrogates_tact,gradient_parcellated).flatten()
test_stat = stats.pearsonr(t_act,gradient_parcellated)[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[0,4], pmat[0,4] = test_stat, pval 
# 
#### For FC
surrogate_brainmap_corrs = pearsonr(surrogates_fc,t_act).flatten()
test_stat = stats.pearsonr(np.mean(fcchange,axis=1),t_act)[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[1,0], pmat[1,0] = test_stat, pval
#
surrogate_brainmap_corrs = pearsonr(surrogates_fc,np.nanmean(taus_rest,axis=1)).flatten()
test_stat = stats.pearsonr(np.mean(fcchange,axis=1),np.nanmean(taus_rest,axis=1))[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[1,2], pmat[1,2] = test_stat, pval
#
surrogate_brainmap_corrs = pearsonr(surrogates_fc,myelinmap_parc).flatten()
test_stat = stats.pearsonr(np.mean(fcchange,axis=1),myelinmap_parc)[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[1,3], pmat[1,3] = test_stat, pval
#
surrogate_brainmap_corrs = pearsonr(surrogates_fc,gradient_parcellated).flatten()
test_stat = stats.pearsonr(np.mean(fcchange,axis=1),gradient_parcellated)[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[1,4], pmat[1,4] = test_stat, pval
#
#### For timescale
surrogate_brainmap_corrs = pearsonr(surrogates_timescale,t_act).flatten()
test_stat = stats.pearsonr(np.nanmean(taus_rest,axis=1),t_act)[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[2,0], pmat[2,0] = test_stat, pval
#
surrogate_brainmap_corrs = pearsonr(surrogates_timescale,np.mean(fcchange,axis=1)).flatten()
test_stat = stats.pearsonr(np.nanmean(taus_rest,axis=1),np.mean(fcchange,axis=1))[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[2,1], pmat[2,1] = test_stat, pval
#
surrogate_brainmap_corrs = pearsonr(surrogates_timescale,myelinmap_parc).flatten()
test_stat = stats.pearsonr(np.nanmean(taus_rest,axis=1),myelinmap_parc)[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[2,3], pmat[2,3] = test_stat, pval
# 
surrogate_brainmap_corrs = pearsonr(surrogates_timescale,gradient_parcellated).flatten()
test_stat = stats.pearsonr(np.nanmean(taus_rest,axis=1),gradient_parcellated)[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[2,4], pmat[2,4] = test_stat, pval
#
#### For myelin map
surrogate_brainmap_corrs = pearsonr(surrogates_myelin,t_act).flatten()
test_stat = stats.pearsonr(myelinmap_parc,t_act)[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[3,0], pmat[3,0] = test_stat, pval
#
surrogate_brainmap_corrs = pearsonr(surrogates_myelin,np.mean(fcchange,axis=1)).flatten()
test_stat = stats.pearsonr(myelinmap_parc,np.mean(fcchange,axis=1))[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[3,1], pmat[3,1] = test_stat, pval
#
surrogate_brainmap_corrs = pearsonr(surrogates_myelin,np.nanmean(taus_rest,axis=1)).flatten()
test_stat = stats.pearsonr(myelinmap_parc,np.nanmean(taus_rest,axis=1))[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[3,2], pmat[3,2] = test_stat, pval
#
surrogate_brainmap_corrs = pearsonr(surrogates_myelin,gradient_parcellated).flatten()
test_stat = stats.pearsonr(myelinmap_parc,gradient_parcellated)[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[3,4], pmat[3,4] = test_stat, pval
#
#### For principal gradient
surrogate_brainmap_corrs = pearsonr(surrogates_gradient,t_act).flatten()
test_stat = stats.pearsonr(gradient_parcellated,t_act)[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[4,0], pmat[4,0] = test_stat, pval
# 
surrogate_brainmap_corrs = pearsonr(surrogates_gradient,np.mean(fcchange,axis=1)).flatten()
test_stat = stats.pearsonr(gradient_parcellated,np.mean(fcchange,axis=1))[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[4,1], pmat[4,1] = test_stat, pval
#
surrogate_brainmap_corrs = pearsonr(surrogates_gradient,np.nanmean(taus_rest,axis=1)).flatten()
test_stat = stats.pearsonr(gradient_parcellated,np.nanmean(taus_rest,axis=1))[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[4,2], pmat[4,2] = test_stat, pval
# 
surrogate_brainmap_corrs = pearsonr(surrogates_gradient,myelinmap_parc).flatten()
test_stat = stats.pearsonr(gradient_parcellated,myelinmap_parc)[0]
pval = nonparp(test_stat, surrogate_brainmap_corrs)
print("SA-corrected p-value:", pval)
corrmat[4,3], pmat[4,3] = test_stat, pval
np.fill_diagonal(corrmat,np.nan)

ind = [0,1,2,3,4]
labels = np.asarray(labels)[ind]

ind = np.asarray(ind,dtype=int)
ind.shape = (len(ind),1)
plt.figure(figsize=(8,8))
sns.set(font_scale=1.75)
ax = sns.heatmap(corrmat[ind,ind.T],square=True,center=0,vmin=-0.6,vmax=0.6,
                 cmap='RdBu_r',cbar_kws={'fraction':0.046},
                 annot=True,annot_kws={"size": 24})
ax.invert_yaxis()
plt.title('Standard task GLM approach\nTimepoint-to-timepoint FC',fontname='Arial',fontsize=30)
# plt.xlabel('Cortical properties',fontname='Arial',fontsize=24)
# plt.ylabel('Cortical properties',fontname='Arial',fontsize=24)
plt.xticks(np.arange(0.5,5.5),labels,fontsize=22,rotation=-45)
plt.yticks(np.arange(0.5,5.5),labels,fontsize=22,rotation=0)
plt.tight_layout()
# plt.savefig('../figures/manuscriptNetDim2b/fMRI4_SummaryFigure_TaskGLM_Replication.pdf')

# compute fdr-corrected q-values
triu_ind = np.triu_indices(5,1)
qarr = mc.fdrcorrection0(pmat[triu_ind])[1]
print('Maximum FDR-corrected q:', np.max(qarr))
```
