```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
import statsmodels.sandbox.stats.multicomp as mc
import seaborn as sns
plt.style.use("ggplot")
sns.set_style('whitegrid')
plt.rcParams["font.family"] = "Arial"
import nibabel as nib
import pandas as pd
import h5py
from importlib import reload
import os
```

## Set-up subject list and construct global variables

```python
## General parameters/variables nParcels = 360

glasserfile2 = '../../data/Q1-Q6_RelatedParcellation210.LR.CorticalAreas_dil_Colors.32k_fs_RL.dlabel.nii'
glasser2 = nib.load(glasserfile2).get_data()
glasser2 = np.squeeze(glasser2)

color1, color2 = sns.color_palette(palette='pastel',n_colors=2)


## General parameters/variables
nParcels = 360

glasserfile2 = '/projects/f_mc1689_1/AnalysisTools/ParcelsGlasser2016/Q1-Q6_RelatedParcellation210.LR.CorticalAreas_dil_Colors.32k_fs_RL.dlabel.nii'
glasserfile2 = '../../data/Q1-Q6_RelatedParcellation210.LR.CorticalAreas_dil_Colors.32k_fs_RL.dlabel.nii'
glasser2 = nib.load(glasserfile2).get_data()
glasser2 = np.squeeze(glasser2)


```

# Set association/sensorimotor network palette and map to surface

```python
# Using final partition
networkdef = np.loadtxt('../../data/network_partition.txt')
right_rh = networkdef[180:]
left_rh = networkdef[:180]
networkdef = np.hstack((right_rh,left_rh))
networkorder = np.asarray(sorted(range(len(networkdef)), key=lambda k: networkdef[k]))
networkorder.shape = (len(networkorder),1)
# network mappings for final partition set
networkmappings = {'fpn':7, 'vis1':1, 'vis2':2, 'smn':3, 'aud':8, 'lan':6, 'dan':5, 'con':4, 'dmn':9, 
                   'pmulti':10, 'none1':11, 'none2':12}
networks = networkmappings.keys()

associationnets = ['con','dan','lan','fpn','dmn','pmulti','none1','none2']
sensorimotornets = ['vis1','vis2','smn','aud']

roi_id = np.zeros((nParcels,))
tmp = {}
netcount = 0
for network in networkmappings:
    net = np.where(networkdef==networkmappings[network])[0]
    for roi in net:
        if network in associationnets:
            tmp[roi] = color1 #'r'
            roi_id[roi] = -1
        else:
            tmp[roi] = color2 #'b'
            roi_id[roi] = 1
    netcount += 1
    
roiColorsByNetwork = []
for roi in range(len(networkdef)):
    roiColorsByNetwork.append(tmp[roi])
 
    
roi_64k = np.zeros((glasser2.shape[0],)) # fc diff, t, p, q
for roi in range(nParcels):
    roi_ind = np.where(glasser2==roi+1)[0]
    
    roi_64k[roi_ind] = roi_id[roi]
    
# Compute effect size baseline (information content)
outdir = '../../data/results/'
filename = 'Networkmapping_SensoryAssociation'
np.savetxt(outdir + filename + '.csv', roi_64k,fmt='%s')
wb_command = 'wb_command -cifti-convert -from-text ' + outdir + filename + '.csv ' + glasserfile2 + ' ' + outdir + filename + '.dscalar.nii -reset-scalars'
os.system(wb_command)
```

# Load in result data


#### Load in task activity and connectivity data

```python
resultdir = '../../data/results/'
#
activity = np.loadtxt(resultdir + '/activity/activity_canonicalHRF.csv',delimiter=',')
t_act = np.mean(activity,axis=1)
subj_activity = np.loadtxt(resultdir + '/activity/activity_canonicalHRF_allSubjs.csv',delimiter=',')

connectivity = np.loadtxt(resultdir + '/connectivity/conn_taskVrest_gbc_FIR.csv', delimiter=',')

nsubjs = subj_activity.shape[1]
```

#### Load in myelin data

```python
myelinmap = np.squeeze(nib.load('../../data/results/Mean.339.MyelinMap_BC_MSMAll.32k_fs_LR.dscalar.nii').get_data())
nvertices = int(glasser2.shape[0]/2)

glasser_lr = np.hstack((glasser2[nvertices:],glasser2[:nvertices])) # need to flip this original parcellation

myelinmap_parc = np.zeros((nParcels,))
for roi in range(nParcels):
    roi_ind = np.where(glasser_lr==roi+1)[0]
    myelinmap_parc[roi] = np.mean(myelinmap[roi_ind])
```

#### Load in timescale data

```python
# Load data
taus_rest = np.loadtxt('../../data/results/timescales/Murray_taus_regions_autocorrelation_rest100.txt',delimiter=',')
As_rest = np.loadtxt('../../data/results/timescales/Murray_As_regions_autocorrelation_rest100.txt',delimiter=',')
Bs_rest = np.loadtxt('../../data/results/timescales/Murray_Bs_regions_autocorrelation_rest100.txt',delimiter=',')
bad_ind = np.where(taus_rest>50)
taus_rest[bad_ind] = np.nan
```

#### Load actflow data

```python
h5f = h5py.File('../../data/results/actflow/activations_actflow_analysis.h5','r')
task_activations = h5f['real'][:].copy()
actflow_activations = h5f['actflow'][:].copy()
h5f.close()

group_task_act = np.mean(task_activations,axis=2)
group_task_actflow = np.mean(actflow_activations,axis=2)

pred_act_sim = np.zeros((nParcels,))
for roi in range(group_task_act.shape[0]):
    pred_act_sim[roi] = np.mean(np.abs(group_task_act[roi,:]-group_task_actflow[roi,:]))

subj_task_act = np.mean(np.abs(task_activations-actflow_activations),axis=1)
```

# Activity X Connectivity data

```python
## Activity decodability
sensorimotor_association = np.zeros((2,nsubjs))

for net in associationnets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[0,:] = sensorimotor_association[0,:] + np.nanmean(subj_activity[net_ind,:],axis=0)
sensorimotor_association[0,:] = sensorimotor_association[0,:]/len(associationnets)

for net in sensorimotornets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[1,:] = sensorimotor_association[1,:] + np.nanmean(subj_activity[net_ind,:],axis=0)
sensorimotor_association[1,:] = sensorimotor_association[1,:]/len(sensorimotornets)

plt.figure(figsize=(5,7))
sns.boxplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
           meanline=True,showmeans=True,meanprops={'linewidth':5,'color':'k','linestyle':':'},
           medianprops={'linestyle':'-'},sym='',whis=0)
sns.swarmplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
            alpha=0.5)
plt.title('Average task\nactivation',fontsize=30,fontname='Arial')
plt.ylabel('Activation (a.u.)',fontsize=26)
plt.xlabel('Network',fontsize=26)
plt.xticks(np.arange(2),['Transmodal','Unimodal'],rotation=0,fontname='Arial',fontsize=18)
plt.yticks(fontsize=18)
plt.tight_layout()
# plt.savefig('../figures/manuscriptNetDim2b/fMRI2s_TransmodalVUnimodal_ActivityChange_Set2.pdf')
print('Activity change')
print('Activity difference:', np.mean(sensorimotor_association[0,:])-np.mean(sensorimotor_association[1,:]))
print(stats.ttest_rel(sensorimotor_association[0,:],sensorimotor_association[1,:]))


##Connectivity decodability
sensorimotor_association = np.zeros((2,nsubjs))

for net in associationnets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[0,:] = sensorimotor_association[0,:] + np.nanmean(connectivity[net_ind,:],axis=0)
sensorimotor_association[0,:] = sensorimotor_association[0,:]/len(associationnets)

for net in sensorimotornets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[1,:] = sensorimotor_association[1,:] + np.nanmean(connectivity[net_ind,:],axis=0)
sensorimotor_association[1,:] = sensorimotor_association[1,:]/len(sensorimotornets)

plt.figure(figsize=(5,7))
sns.boxplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
           meanline=True,showmeans=True,meanprops={'linewidth':5,'color':'k','linestyle':':'},
           medianprops={'linestyle':'-'},sym='',whis=0)
sns.swarmplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
            alpha=0.5)
plt.title('Average FC\nchange',fontsize=30,fontname='Arial')
plt.ylabel('FC change (r)',fontsize=26)
plt.xlabel('Network',fontsize=26)
plt.xticks(np.arange(2),['Transmodal','Unimodal'],rotation=0,fontname='Arial',fontsize=18)
plt.yticks(fontsize=18)
plt.tight_layout()
# plt.savefig('../figures/manuscriptNetDim2b/fMRI2s_TransmodalVUnimodal_FCChange_Set2.pdf')
print('FC change')
print('FC difference:', np.mean(sensorimotor_association[0,:])-np.mean(sensorimotor_association[1,:]))
print(stats.ttest_rel(sensorimotor_association[0,:],sensorimotor_association[1,:]))
```

# Calculate local time scale from rest data (Murray approach)

```python
##Intrinsic timescale
sensorimotor_association = np.zeros((2,nsubjs))

for net in associationnets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[0,:] = sensorimotor_association[0,:] + np.nanmean(taus_rest[net_ind,:],axis=0)
sensorimotor_association[0,:] = sensorimotor_association[0,:]/len(associationnets)

for net in sensorimotornets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[1,:] = sensorimotor_association[1,:] + np.nanmean(taus_rest[net_ind,:],axis=0)
sensorimotor_association[1,:] = sensorimotor_association[1,:]/len(sensorimotornets)

plt.figure(figsize=(5,7))
sns.boxplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
           meanline=True,showmeans=True,meanprops={'linewidth':5,'color':'k','linestyle':':'},
           medianprops={'linestyle':'-'},sym='',whis=0)
sns.swarmplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
            alpha=0.5)
plt.title('Intrinsic timescale',fontsize=30,fontname='Arial')
plt.ylabel(r'Intrinsic timescale ($\tau$)',fontsize=26)
plt.xlabel('Network',fontsize=26)
plt.xticks(np.arange(2),['Transmodal','Unimodal'],rotation=0,fontname='Arial',fontsize=24)
plt.yticks(fontsize=24)
plt.tight_layout()
# plt.savefig('../figures/manuscriptNetDim2b/fMRI2s_TransVUnimodal_IntrinsicTimescale_Set2.pdf')
print('Intrinsic timescale')
print('Average tau diff', np.mean(sensorimotor_association[0,:]) - np.mean(sensorimotor_association[1,:]))
print(stats.ttest_rel(sensorimotor_association[0,:],sensorimotor_association[1,:]))
```

#### Plot autocorrelation function with longest/shortest tau

```python
# Find fastest/slowest region
fast = np.where(np.nanmean(taus_rest,axis=1)==np.min(np.nanmean(taus_rest,axis=1)))[0][0]
# slow = np.where(np.nanmean(taus_rest,axis=1)==np.max(np.nanmean(taus_rest,axis=1)))[0][0]
slow = 271
# fast = 73 + 180
# slow = 3
ntps = 20
df_slow = {}
df_fast = {}

df_slow['Subject'] = []
df_slow['Timepoint'] = []
df_slow['Autocorrelation'] = []

df_fast['Subject'] = []
df_fast['Timepoint'] = []
df_fast['Autocorrelation'] = []

mean_fast = np.zeros((nsubjs,ntps))
mean_slow = np.zeros((nsubjs,ntps))
for s in range(nsubjs):
    for tp in range(ntps):
        df_fast['Subject'].append(s)
        df_fast['Timepoint'].append(tp)
        y = As_rest[fast,s]*(np.exp(-tp/taus_rest[fast,s])+Bs_rest[fast,s])
        df_fast['Autocorrelation'].append(y)
        mean_fast[s,tp] = y
        
        df_slow['Subject'].append(s)
        df_slow['Timepoint'].append(tp)
        y = As_rest[slow,s]*(np.exp(-tp/taus_rest[slow,s])+Bs_rest[slow,s])
        df_slow['Autocorrelation'].append(y)
        mean_slow[s,tp] = y
        
df_fast = pd.DataFrame(df_fast)
df_slow = pd.DataFrame(df_slow)
```

```python

plt.figure(figsize=(10,7));
sns.regplot(x="Timepoint", y="Autocorrelation", data=df_fast, color='#add8e6', label='Fast, Parcel ' + str(fast), x_estimator=np.mean,fit_reg=False);
sns.regplot(x="Timepoint", y="Autocorrelation", data=df_slow, color='#f08080', label='Slow, Parcel ' + str(slow), x_estimator=np.mean,fit_reg=False);
x = np.arange(ntps)
# plt.plot(x,np.mean(mean_fast,axis=0),color='#add8e6',linewidth=3,label='Slow')
# plt.plot(x,np.mean(mean_slow,axis=0),color='#f08080',linewidth=3,label='Fast')
# plt.xticks(np.arange(ntps),np.arange(ntps),10);
plt.legend(prop={'size': 24});
plt.xlabel('Time points (TRs)',fontsize=26);
plt.ylabel('Autocorrelation',fontsize=26);
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.title('Autocorrelation function\nExample ROIs',fontsize=30);
plt.tight_layout()
# plt.savefig('../figures/manuscriptNetDim2b/fMRI2s_AutocorrelationExample_Set2.pdf')
```
