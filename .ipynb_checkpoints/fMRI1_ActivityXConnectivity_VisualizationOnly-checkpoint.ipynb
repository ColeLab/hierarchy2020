# Load in result data


```python
resultdir = '../../data/results/'
decod_act = np.loadtxt(resultdir + 'act_decoding_acc.csv', delimiter=',')
decod_con = np.loadtxt(resultdir + 'conn_decoding_acc_maxTPs.csv', delimiter=',')
dimen_act = pd.read_csv(resultdir + 'act_dim_corr_dataframe.csv')
dimen_con = pd.read_csv(resultdir + 'conn_dim_corr_dataframe_maxTPs.csv')
```

# Decodability results for connectivity and activity

```python
connect_dec = np.zeros((len(orderednetworks,)))
activit_dec = np.zeros((len(orderednetworks,)))
i = 0
for net in orderednetworks:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    connect_dec[i] = np.mean(decod_con[net_ind,:])
    activit_dec[i] = np.mean(decod_act[net_ind,:])
    i += 1
```

```python
plt.figure(figsize=(14,5))
plt.subplot(121)
plt.bar(np.arange(len(connect_dec)),connect_dec)
plt.xticks(np.arange(len(connect_dec)),orderednetworks)
# plt.ylim([1.25,2])
plt.title('Connectivity Decodability');

plt.subplot(122)
plt.bar(np.arange(len(activit_dec)),activit_dec)
plt.xticks(np.arange(len(activit_dec)),orderednetworks)
plt.title('Activity Decodability');

```

## Run correlation analysis (group-level and within-subject) comparing activity x connectivity decodability

#### Group-level correlation

```python
stats.spearmanr(np.mean(decod_act,axis=1),np.mean(decod_con,axis=1))
```

```python
networkpalette = ['royalblue','slateblue','paleturquoise','darkorchid','limegreen',
                  'lightseagreen','yellow','orchid','r','peru','orange','olivedrab']
networkpalette = np.asarray(networkpalette)
tmp = {}
netcount = 0
for network in orderednetworks:
    net = np.where(networkdef==networkmappings[network])[0]
    for roi in net:
        tmp[roi] = networkpalette[netcount]
    netcount += 1

roiColorsByNetwork = []
for roi in range(len(networkdef)):
    roiColorsByNetwork.append(tmp[roi])
    

plt.figure(figsize=(10,7))
sns.regplot(np.mean(decod_act,axis=1)*100.0,np.mean(decod_con,axis=1)*100.0,color='k')
sns.regplot(np.mean(decod_act,axis=1)*100.0,np.mean(decod_con,axis=1)*100.0,color='k',scatter_kws={'color':roiColorsByNetwork})
plt.ylabel('Connectivity decoding accuracy (%)',fontsize=24)
plt.xlabel('Activity decoding accuracy (%)',fontsize=24)
plt.title('Decodability of task activations and connectivity\nare negatively correlated',fontsize=25)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.tight_layout()
# plt.savefig('../figures/2020OHBM/fMRI1_ActivityXConnectivity_Decoding_Exploratory.pdf')
```

#### Within-subject correlation

```python
nsubjs = decod_act.shape[1]
decoding_corr = []
for s in range(nsubjs):
    decoding_corr.append(stats.spearmanr(decod_act[:,s],decod_con[:,s])[0])
    
t, p = stats.ttest_1samp(decoding_corr,0)
print('Average correlation, r =', np.mean(decoding_corr))
print('t =', t, '| p =', p)
```

# Dimensionality results for connectivity and activity

```python
dimen_conn2d = np.zeros((nParcels,len(subjNums)))
dimen_act2d = np.zeros((nParcels,len(subjNums)))
for parcel in range(nParcels):
    dimen_conn2d[parcel,:] = dimen_con.loc[dimen_con['Parcel']==parcel+1].Dimensionality
    dimen_act2d[parcel,:] = dimen_act.loc[dimen_act['Parcel']==parcel+1].Dimensionality
```

```python
connect_dim = np.zeros((len(orderednetworks,)))
activit_dim = np.zeros((len(orderednetworks,)))
i = 0
for net in orderednetworks:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    connect_dim[i] = np.mean(dimen_conn2d[net_ind,:])
    activit_dim[i] = np.mean(dimen_act2d[net_ind,:])
    i += 1
```

```python
plt.figure(figsize=(14,5))
plt.subplot(121)
plt.bar(np.arange(len(connect_dim)),connect_dim)
plt.xticks(np.arange(len(connect_dim)),orderednetworks)
# plt.ylim([1.25,2])
plt.title('Connectivity Dimensionality');

plt.subplot(122)
plt.bar(np.arange(len(activit_dim)),activit_dim)
plt.xticks(np.arange(len(activit_dim)),orderednetworks)
plt.title('Activity Dimensionality');

```

```python
# Load data
timescales_rest = np.loadtxt('../../data/results/timescale_regions_autocorrelation_rest.txt',delimiter=',')
network_tscale = np.zeros((len(orderednetworks),len(subjNums)))
i = 0
for net in orderednetworks:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    network_tscale[i,:] = np.mean(timescales_rest[net_ind,:],axis=0)
    i += 1
    
avg = np.mean(network_tscale,axis=1)
sorted_index = sorted(range(len(orderednetworks)), key=lambda k: avg[k])

plt.figure(figsize=(7,5))
sns.swarmplot(data=network_tscale[sorted_index,:].T, 
               palette=networkpalette[sorted_index],alpha=.5)
sns.boxplot(data=network_tscale[sorted_index,:].T, 
               palette=networkpalette[sorted_index],whis=np.inf)
plt.plot(range(len(orderednetworks[:])),np.repeat(np.mean(network_tscale),len(orderednetworks[:])),'k--',linewidth=4)
plt.xticks(np.arange(len(OrderedNetworks[:])),np.asarray(OrderedNetworks)[sorted_index], fontsize=16, fontname='FreeSans', rotation=-45)
plt.xlabel('Network',fontname='FreeSans', labelpad=20, fontsize=20)
plt.ylabel('Timescale',fontname='FreeSans',fontsize=20)
plt.yticks(fontsize=16,fontname='FreeSans')
plt.title('Regional timescales, Rest', y=1.02,fontsize=24, fontname='FreeSans')
plt.tight_layout()
# plt.savefig('Glasser6_Exploratory_ConnDim_CV.pdf')
```

```python
# Load data
taus_rest = np.loadtxt('../../data/results/Murray_taus_regions_autocorrelation_rest.txt',delimiter=',')
As_rest = np.loadtxt('../../data/results/Murray_As_regions_autocorrelation_rest.txt',delimiter=',')
Bs_rest = np.loadtxt('../../data/results/Murray_Bs_regions_autocorrelation_rest.txt',delimiter=',')
bad_ind = np.where(taus_rest>50)
taus_rest[bad_ind] = np.nan
# bad_ind = np.where(taus_rest<0)
# taus_rest[bad_ind] = np.nan


network_tscale = np.zeros((len(orderednetworks),len(subjNums)))
i = 0
for net in orderednetworks:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    network_tscale[i,:] = np.nanmean(taus_rest[net_ind,:],axis=0)
    i += 1
    
avg = np.mean(network_tscale,axis=1)
sorted_index = sorted(range(len(orderednetworks)), key=lambda k: avg[k])

plt.figure(figsize=(7,5))
sns.swarmplot(data=network_tscale[sorted_index,:].T, 
               palette=networkpalette[sorted_index],alpha=.5)
sns.boxplot(data=network_tscale[sorted_index,:].T, 
               palette=networkpalette[sorted_index],whis=np.inf)
plt.plot(range(len(orderednetworks[:])),np.repeat(np.mean(network_tscale),len(orderednetworks[:])),'k--',linewidth=4)
plt.xticks(np.arange(len(OrderedNetworks[:])),np.asarray(OrderedNetworks)[sorted_index], fontsize=16, fontname='FreeSans', rotation=-45)
plt.xlabel('Network',fontname='FreeSans', labelpad=20, fontsize=20)
plt.ylabel('Timescale',fontname='FreeSans',fontsize=20)
plt.yticks(fontsize=16,fontname='FreeSans')
plt.title('Regional timescales, Rest', y=1.02,fontsize=24, fontname='FreeSans')
plt.tight_layout()
# plt.savefig('Glasser6_Exploratory_ConnDim_CV.pdf')
```
