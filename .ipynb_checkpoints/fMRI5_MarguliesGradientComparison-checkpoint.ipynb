# Load in Margulies data and parcellate, and create parcellated Margulies output

```python
#######################################################################################################
#### Load gradient data
gradient_rh = np.squeeze(nib.load('../../data/results/margulies2016/hcp.tmp.rh.dscalar.nii').get_data())
gradient_lh = np.squeeze(nib.load('../../data/results/margulies2016/hcp.tmp.lh.dscalar.nii').get_data())
# Glasser atlas is orgnized R->L hemispheres
gradient = np.hstack((gradient_lh,gradient_rh))
#######################################################################################################

gradient_parcellated = np.zeros((nParcels,))
for roi in range(nParcels):
    roi_ind = np.where(glasser2==roi+1)[0]
    gradient_parcellated[roi] = np.mean(gradient[roi_ind])

#######################################################################################################
# Remap parcellated data on the cortical surface
parcel2surf = np.zeros((glasser2.shape[0],1)) # fc diff, t, p, q

for roi in range(nParcels):
    roi_ind = np.where(glasser2==roi+1)[0]
    parcel2surf[roi_ind,0] = np.mean(gradient_parcellated[roi])
    
#######################################################################################################
# # Compute effect size baseline (information content)
# outdir = '../../data/results/'
# filename = 'MarguliesGradientParcellated'
# np.savetxt(outdir + filename + '.csv', parcel2surf,fmt='%s')
# wb_command = 'wb_command -cifti-convert -from-text ' + outdir + filename + '.csv ' + glasserfile2 + ' ' + outdir + filename + '.dscalar.nii -reset-scalars'
# os.system(wb_command)
```

# Activity and Connectivity Change results 

```python
connect_dec = np.zeros((len(orderednetworks,)))
activit_dec = np.zeros((len(orderednetworks,)))
i = 0
for net in orderednetworks:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    connect_dec[i] = np.mean(decod_con[net_ind,:])
    activit_dec[i] = np.mean(decod_act[net_ind,:])
    i += 1
```

```python
## Activity decodability
sensorimotor_association = np.zeros((2,len(subjNums)))

for net in associationnets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[0,:] = sensorimotor_association[0,:] + np.nanmean(decod_act[net_ind,:],axis=0)
sensorimotor_association[0,:] = sensorimotor_association[0,:]/len(associationnets)

for net in sensorimotornets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[1,:] = sensorimotor_association[1,:] + np.nanmean(decod_act[net_ind,:],axis=0)
sensorimotor_association[1,:] = sensorimotor_association[1,:]/len(sensorimotornets)

plt.figure(figsize=(5,7))
sns.boxplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
           meanline=True,showmeans=True,meanprops={'linewidth':5,'color':'k','linestyle':':'},
           medianprops={'linestyle':'-'},sym='',whis=0)
sns.swarmplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
            alpha=0.5)
plt.title('Average task\nactivation',fontsize=30,fontname='Arial')
plt.ylabel('Activation (a.u.)',fontsize=26)
plt.xlabel('Network',fontsize=26)
plt.xticks(np.arange(2),['Transmodal','Unimodal'],rotation=0,fontname='Arial',fontsize=18)
plt.yticks(fontsize=18)
plt.tight_layout()
# #plt.savefig('../figures/manuscriptNetDim2b/fMRI2s_TransmodalVUnimodal_ActivityChange_Set2.pdf')
print('Activity change')
print('Activity difference:', np.mean(sensorimotor_association[0,:])-np.mean(sensorimotor_association[1,:]))
print(stats.ttest_rel(sensorimotor_association[0,:],sensorimotor_association[1,:]))


##Connectivity decodability
sensorimotor_association = np.zeros((2,len(subjNums)))

for net in associationnets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[0,:] = sensorimotor_association[0,:] + np.nanmean(decod_con[net_ind,:],axis=0)
sensorimotor_association[0,:] = sensorimotor_association[0,:]/len(associationnets)

for net in sensorimotornets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[1,:] = sensorimotor_association[1,:] + np.nanmean(decod_con[net_ind,:],axis=0)
sensorimotor_association[1,:] = sensorimotor_association[1,:]/len(sensorimotornets)

plt.figure(figsize=(5,7))
sns.boxplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
           meanline=True,showmeans=True,meanprops={'linewidth':5,'color':'k','linestyle':':'},
           medianprops={'linestyle':'-'},sym='',whis=0)
sns.swarmplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
            alpha=0.5)
plt.title('Average FC\nchange',fontsize=30,fontname='Arial')
plt.ylabel('FC change (r)',fontsize=26)
plt.xlabel('Network',fontsize=26)
plt.xticks(np.arange(2),['Transmodal','Unimodal'],rotation=0,fontname='Arial',fontsize=18)
plt.yticks(fontsize=18)
plt.tight_layout()
# #plt.savefig('../figures/manuscriptNetDim2b/fMRI2s_TransmodalVUnimodal_FCChange_Set2.pdf')
print('FC change')
print('FC difference:', np.mean(sensorimotor_association[0,:])-np.mean(sensorimotor_association[1,:]))
print(stats.ttest_rel(sensorimotor_association[0,:],sensorimotor_association[1,:]))
```

## Run correlation analysis (group-level) comparing activity and connectivity decodability with Margulies gradient

#### Group-level correlation

```python
t_act = np.mean(np.abs(stats.ttest_1samp(decod_act_alltasks,0,axis=2)[0]),axis=1)
print('Spearman:',stats.spearmanr(t_act,gradient_parcellated))
print('Pearson:', stats.pearsonr(t_act,gradient_parcellated))

plt.figure(figsize=(10,7))
sns.regplot(t_act,gradient_parcellated,color='k',fit_reg=False)
sns.regplot(t_act,gradient_parcellated,color='k',scatter_kws={'color':roiColorsByNetwork})
plt.ylabel('FC change (r)',fontsize=26)
plt.xlabel('Task activation (abs. t-value)',fontsize=26)
plt.title('Task activations and principal gradient\nare positively correlated',fontsize=30)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.tight_layout()
# #plt.savefig('../figures/manuscriptNetDim2b/fMRI2s_ActivityXConnectivity_Univariate_Set2.pdf')
```

# Calculate local time scale from rest data (Murray approach)

```python
# Load data
taus_rest = np.loadtxt('../../data/results/Murray_taus_regions_autocorrelation_rest100.txt',delimiter=',')
As_rest = np.loadtxt('../../data/results/Murray_As_regions_autocorrelation_rest100.txt',delimiter=',')
Bs_rest = np.loadtxt('../../data/results/Murray_Bs_regions_autocorrelation_rest100.txt',delimiter=',')
bad_ind = np.where(taus_rest>50)
taus_rest[bad_ind] = np.nan
# bad_ind = np.where(taus_rest<0)
# taus_rest[bad_ind] = np.nan


network_tscale = np.zeros((len(orderednetworks),len(subjNums)))
i = 0
for net in orderednetworks:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    network_tscale[i,:] = np.nanmean(taus_rest[net_ind,:],axis=0)
    i += 1
    
avg = np.mean(network_tscale,axis=1)
sorted_index = sorted(range(len(orderednetworks)), key=lambda k: avg[k])

plt.figure(figsize=(7,5))
sns.swarmplot(data=network_tscale[sorted_index,:].T, 
               palette=networkpalette[sorted_index],alpha=.5)
sns.boxplot(data=network_tscale[sorted_index,:].T, 
               palette=networkpalette[sorted_index],whis=np.inf)
plt.plot(range(len(orderednetworks[:])),np.repeat(np.mean(network_tscale),len(orderednetworks[:])),'k--',linewidth=4)
plt.xticks(np.arange(len(OrderedNetworks[:])),np.asarray(OrderedNetworks)[sorted_index], fontsize=16, fontname='Arial', rotation=-45)
plt.xlabel('Network',fontname='Arial', labelpad=20, fontsize=20)
plt.ylabel('Timescale',fontname='Arial',fontsize=20)
plt.yticks(fontsize=16,fontname='Arial')
plt.title('Regional timescales, Rest', y=1.02,fontsize=24, fontname='Arial')
plt.tight_layout()
# #plt.savefig('Glasser6_Exploratory_ConnDim_CV.pdf')
```

```python
##Intrinsic timescale
sensorimotor_association = np.zeros((2,len(subjNums)))

for net in associationnets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[0,:] = sensorimotor_association[0,:] + np.nanmean(taus_rest[net_ind,:],axis=0)
sensorimotor_association[0,:] = sensorimotor_association[0,:]/len(associationnets)

for net in sensorimotornets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[1,:] = sensorimotor_association[1,:] + np.nanmean(taus_rest[net_ind,:],axis=0)
sensorimotor_association[1,:] = sensorimotor_association[1,:]/len(sensorimotornets)

plt.figure(figsize=(5,7))
sns.boxplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
           meanline=True,showmeans=True,meanprops={'linewidth':5,'color':'k','linestyle':':'},
           medianprops={'linestyle':'-'},sym='',whis=0)
sns.swarmplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
            alpha=0.5)
plt.title('Intrinsic timescale',fontsize=30,fontname='Arial')
plt.ylabel(r'Intrinsic timescale ($\tau$)',fontsize=26)
plt.xlabel('Network',fontsize=26)
plt.xticks(np.arange(2),['Transmodal','Unimodal'],rotation=0,fontname='Arial',fontsize=24)
plt.yticks(fontsize=24)
plt.tight_layout()
#plt.savefig('../figures/manuscriptNetDim2b/fMRI2s_TransVUnimodal_IntrinsicTimescale_Set2.pdf')
print('Intrinsic timescale')
print('Average tau diff', np.mean(sensorimotor_association[0,:]) - np.mean(sensorimotor_association[1,:]))
print(stats.ttest_rel(sensorimotor_association[0,:],sensorimotor_association[1,:]))
```

#### Plot autocorrelation function with longest/shortest tau

```python
# Find fastest/slowest region
fast = np.where(np.nanmean(taus_rest,axis=1)==np.min(np.nanmean(taus_rest,axis=1)))[0][0]
slow = np.where(np.nanmean(taus_rest,axis=1)==np.max(np.nanmean(taus_rest,axis=1)))[0][0]
# fast = 73 + 180
# slow = 3
ntps = 20
df_slow = {}
df_fast = {}

df_slow['Subject'] = []
df_slow['Timepoint'] = []
df_slow['Autocorrelation'] = []

df_fast['Subject'] = []
df_fast['Timepoint'] = []
df_fast['Autocorrelation'] = []

mean_fast = np.zeros((len(subjNums),ntps))
mean_slow = np.zeros((len(subjNums),ntps))
for s in range(len(subjNums)):
    for tp in range(ntps):
        df_fast['Subject'].append(s)
        df_fast['Timepoint'].append(tp)
        y = As_rest[fast,s]*(np.exp(-tp/taus_rest[fast,s])+Bs_rest[fast,s])
        df_fast['Autocorrelation'].append(y)
        mean_fast[s,tp] = y
        
        df_slow['Subject'].append(s)
        df_slow['Timepoint'].append(tp)
        y = As_rest[slow,s]*(np.exp(-tp/taus_rest[slow,s])+Bs_rest[slow,s])
        df_slow['Autocorrelation'].append(y)
        mean_slow[s,tp] = y
        
df_fast = pd.DataFrame(df_fast)
df_slow = pd.DataFrame(df_slow)
```

```python

plt.figure(figsize=(10,7));
sns.regplot(x="Timepoint", y="Autocorrelation", data=df_fast, color='#add8e6', label='Fast, Parcel ' + str(fast), x_estimator=np.mean,fit_reg=False);
sns.regplot(x="Timepoint", y="Autocorrelation", data=df_slow, color='#f08080', label='Slow, Parcel ' + str(slow), x_estimator=np.mean,fit_reg=False);
x = np.arange(ntps)
# plt.plot(x,np.mean(mean_fast,axis=0),color='#add8e6',linewidth=3,label='Slow')
# plt.plot(x,np.mean(mean_slow,axis=0),color='#f08080',linewidth=3,label='Fast')
# plt.xticks(np.arange(ntps),np.arange(ntps),10);
plt.legend(prop={'size': 24});
plt.xlabel('Time points (TRs)',fontsize=26);
plt.ylabel('Autocorrelation',fontsize=26);
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.title('Autocorrelation function\nExample ROIs',fontsize=30);
plt.tight_layout()
# #plt.savefig('../figures/manuscriptNetDim2b/fMRI2s_AutocorrelationExample_Set2.pdf')
```

# Construct correlation matrix for all values: 
#### activity, connectivity, timescale, myelin

```python
labels = ['Activity','FC change','Timescale','Myelin']
corrmat = np.zeros((4,4))
pmat = np.zeros((4,4))
# For activity
corrmat[0,1], pmat[0,1] = stats.spearmanr(t_act,np.mean(decod_con,axis=1))
corrmat[0,2], pmat[0,2] = stats.spearmanr(t_act,np.nanmean(taus_rest,axis=1))
corrmat[0,3], pmat[0,3] = stats.spearmanr(t_act,myelinmap_parc)
# For FC
corrmat[1,0], pmat[1,0] = stats.spearmanr(np.mean(decod_con,axis=1),t_act)
corrmat[1,2], pmat[1,2] = stats.spearmanr(np.mean(decod_con,axis=1),np.nanmean(taus_rest,axis=1))
corrmat[1,3], pmat[1,3] = stats.spearmanr(np.mean(decod_con,axis=1),myelinmap_parc)
# For timescale
corrmat[2,0], pmat[2,0] = stats.spearmanr(np.nanmean(taus_rest,axis=1),t_act)
corrmat[2,1], pmat[2,1] = stats.spearmanr(np.nanmean(taus_rest,axis=1),np.mean(decod_con,axis=1))
corrmat[2,3], pmat[2,3] = stats.spearmanr(np.nanmean(taus_rest,axis=1),myelinmap_parc)
# For myelin map
corrmat[3,0], pmat[3,0] = stats.spearmanr(myelinmap_parc,t_act)
corrmat[3,1], pmat[3,1] = stats.spearmanr(myelinmap_parc,np.mean(decod_con,axis=1))
corrmat[3,2], pmat[3,2] = stats.spearmanr(myelinmap_parc,np.nanmean(taus_rest,axis=1))
np.fill_diagonal(corrmat,np.nan)

ind = [0,1,2,3]
labels = np.asarray(labels)[ind]

ind = np.asarray(ind,dtype=int)
ind.shape = (len(ind),1)
plt.figure(figsize=(8,8))
sns.set(font_scale=1.75)
ax = sns.heatmap(corrmat[ind,ind.T],square=True,center=0,vmin=-0.6,vmax=0.6,
                 cmap='RdBu_r',cbar_kws={'fraction':0.046},
                 annot=True,annot_kws={"size": 24})
ax.invert_yaxis()
plt.title('Peak activation approach\nBlock-to-block FC',fontname='Arial',fontsize=30)
# plt.xlabel('Cortical properties',fontname='Arial',fontsize=24)
# plt.ylabel('Cortical properties',fontname='Arial',fontsize=24)
plt.xticks(np.arange(0.5,4.5),labels,fontsize=22,rotation=-45)
plt.yticks(np.arange(0.5,4.5),labels,fontsize=22,rotation=0)
plt.tight_layout()
# plt.savefig('../figures/manuscriptNetDim2b/fMRI4_SummaryFigure_NoTaskGLM_Replication.pdf')

# compute fdr-corrected q-values
triu_ind = np.triu_indices(4,1)
qarr = mc.fdrcorrection0(pmat[triu_ind])[1]
print('Maximum FDR-corrected q:', np.max(qarr))
```
