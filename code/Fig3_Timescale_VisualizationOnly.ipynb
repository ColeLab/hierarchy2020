## Set-up parameters

```python
## General parameters/variables 
nParcels = 360

glasserfile2 = '../resultdata/Q1-Q6_RelatedParcellation210.LR.CorticalAreas_dil_Colors.32k_fs_RL.dlabel.nii'
glasser2 = nib.load(glasserfile2).get_data()
glasser2 = np.squeeze(glasser2)

color1, color2 = sns.color_palette(palette='pastel',n_colors=2)

# Import network partition
networkdef = np.loadtxt('../resultdata/network_partition.txt')
# network mappings for final partition set
networkmappings = {'fpn':7, 'vis1':1, 'vis2':2, 'smn':3, 'aud':8, 'lan':6, 'dan':5, 'con':4, 'dmn':9, 
                   'pmulti':10, 'none1':11, 'none2':12}
networks = networkmappings.keys()
 
associationnets = ['con','dan','lan','fpn','dmn','pmulti','none1','none2']
sensorimotornets = ['vis1','vis2','smn','aud']

roi_id = np.zeros((nParcels,))
tmp = {}
netcount = 0
for network in networkmappings:
    net = np.where(networkdef==networkmappings[network])[0]
    for roi in net:
        if network in associationnets:
            tmp[roi] = color1 #'r'
            roi_id[roi] = -1
        else:
            tmp[roi] = color2 #'b'
            roi_id[roi] = 1
    netcount += 1
    
roiColorsByNetwork = []
for roi in range(len(networkdef)):
    roiColorsByNetwork.append(tmp[roi])
```

# Load in result data


#### Load in timescale, task activity and connectivity data

```python
resultdir = '../resultdata/'
# load activity data
activity = np.loadtxt(resultdir + '/activity/activity_canonicalHRF.csv',delimiter=',')
t_act = np.mean(activity,axis=1)
subj_activity = np.loadtxt(resultdir + '/activity/activity_canonicalHRF_allSubjs.csv',delimiter=',')
# load fc data
connectivity = np.loadtxt(resultdir + '/connectivity/conn_taskVrest_gbc_FIR.csv', delimiter=',')
# load timescale data
timescales = np.loadtxt(resultdir + '/timescales/Murray_taus_regions_autocorrelation_rest100.txt',delimiter=',')
# remove outliers (bad estimates) (only 2, but doesn't make a difference to results)
bad_ind = np.where(timescales>50) 
timescales[bad_ind] = np.nan

nsubjs = subj_activity.shape[1]
```

# Calculate local time scale from rest data (Murray approach)

```python
##Intrinsic timescale
sensorimotor_association = np.zeros((2,nsubjs))

for net in associationnets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[0,:] = sensorimotor_association[0,:] + np.nanmean(timescales[net_ind,:],axis=0)
sensorimotor_association[0,:] = sensorimotor_association[0,:]/len(associationnets)

for net in sensorimotornets:
    net_ind = np.where(networkdef==networkmappings[net])[0]
    sensorimotor_association[1,:] = sensorimotor_association[1,:] + np.nanmean(timescales[net_ind,:],axis=0)
sensorimotor_association[1,:] = sensorimotor_association[1,:]/len(sensorimotornets)

plt.figure(figsize=(5,7))
sns.boxplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
           meanline=True,showmeans=True,meanprops={'linewidth':5,'color':'k','linestyle':':'},
           medianprops={'linestyle':'-'},sym='',whis=0)
sns.swarmplot(data=[sensorimotor_association[0,:],
                  sensorimotor_association[1,:]],
            palette=sns.color_palette(palette='pastel',n_colors=2),
            alpha=0.5)
plt.title('Intrinsic timescale',fontsize=30,fontname='Arial')
plt.ylabel(r'Intrinsic timescale ($\tau$)',fontsize=26)
plt.xlabel('Network',fontsize=26)
plt.xticks(np.arange(2),['Transmodal','Unimodal'],rotation=0,fontname='Arial',fontsize=24)
plt.yticks(fontsize=24)
plt.tight_layout()
# plt.savefig('Fig3_TransVUnimodal_IntrinsicTimescale.pdf')
print('Intrinsic timescale')
print('Average tau diff', np.mean(sensorimotor_association[0,:]) - np.mean(sensorimotor_association[1,:]))
print(stats.ttest_rel(sensorimotor_association[0,:],sensorimotor_association[1,:]))
```

#### Plot autocorrelation function with longest/shortest tau

```python
# Find fastest/slowest region
fast = np.where(np.nanmean(timescales,axis=1)==np.min(np.nanmean(timescales,axis=1)))[0][0]
slow = 271
ntps = 20
df_slow = {}
df_fast = {}

df_slow['Subject'] = []
df_slow['Timepoint'] = []
df_slow['Autocorrelation'] = []

df_fast['Subject'] = []
df_fast['Timepoint'] = []
df_fast['Autocorrelation'] = []

mean_fast = np.zeros((nsubjs,ntps))
mean_slow = np.zeros((nsubjs,ntps))
for s in range(nsubjs):
    for tp in range(ntps):
        df_fast['Subject'].append(s)
        df_fast['Timepoint'].append(tp)
        y = As_rest[fast,s]*(np.exp(-tp/timescales[fast,s])+Bs_rest[fast,s])
        df_fast['Autocorrelation'].append(y)
        mean_fast[s,tp] = y
        
        df_slow['Subject'].append(s)
        df_slow['Timepoint'].append(tp)
        y = As_rest[slow,s]*(np.exp(-tp/timescales[slow,s])+Bs_rest[slow,s])
        df_slow['Autocorrelation'].append(y)
        mean_slow[s,tp] = y
        
df_fast = pd.DataFrame(df_fast)
df_slow = pd.DataFrame(df_slow)
```

```python

plt.figure(figsize=(10,7));
sns.regplot(x="Timepoint", y="Autocorrelation", data=df_fast, color='#add8e6', label='Fast, Parcel ' + str(fast), x_estimator=np.mean,fit_reg=False);
sns.regplot(x="Timepoint", y="Autocorrelation", data=df_slow, color='#f08080', label='Slow, Parcel ' + str(slow), x_estimator=np.mean,fit_reg=False);
x = np.arange(ntps)
# plt.plot(x,np.mean(mean_fast,axis=0),color='#add8e6',linewidth=3,label='Slow')
# plt.plot(x,np.mean(mean_slow,axis=0),color='#f08080',linewidth=3,label='Fast')
# plt.xticks(np.arange(ntps),np.arange(ntps),10);
plt.legend(prop={'size': 24});
plt.xlabel('Time points (TRs)',fontsize=26);
plt.ylabel('Fit to autocorrelation',fontsize=26);
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.title('Exponential fit to\ntemporal autocorrelation function',fontsize=30);
plt.tight_layout()
# plt.savefig('Fig3_AutocorrelationExample.pdf')
```
