## Set-up parameters

```python
## General parameters/variables 
nParcels = 360

glasserfile2 = '../resultdata/Q1-Q6_RelatedParcellation210.LR.CorticalAreas_dil_Colors.32k_fs_RL.dlabel.nii'
glasser2 = nib.load(glasserfile2).get_data()
glasser2 = np.squeeze(glasser2)

color1, color2 = sns.color_palette(palette='pastel',n_colors=2)

# Import network partition
networkdef = np.loadtxt('../resultdata/network_partition.txt')
# network mappings for final partition set
networkmappings = {'fpn':7, 'vis1':1, 'vis2':2, 'smn':3, 'aud':8, 'lan':6, 'dan':5, 'con':4, 'dmn':9, 
                   'pmulti':10, 'none1':11, 'none2':12}
networks = networkmappings.keys()
 
associationnets = ['con','dan','lan','fpn','dmn','pmulti','none1','none2']
sensorimotornets = ['vis1','vis2','smn','aud']

roi_id = np.zeros((nParcels,))
tmp = {}
netcount = 0
for network in networkmappings:
    net = np.where(networkdef==networkmappings[network])[0]
    for roi in net:
        if network in associationnets:
            tmp[roi] = color1 #'r'
            roi_id[roi] = -1
        else:
            tmp[roi] = color2 #'b'
            roi_id[roi] = 1
    netcount += 1
    
roiColorsByNetwork = []
for roi in range(len(networkdef)):
    roiColorsByNetwork.append(tmp[roi])
```

# Load in result data


#### Load in timescale, task activity and connectivity data

```python
resultdir = '../resultdata/'
# load activity data
activity = np.loadtxt(resultdir + '/activity/activity_canonicalHRF_replication.csv',delimiter=',')
t_act = np.mean(activity,axis=1)
subj_activity = np.loadtxt(resultdir + '/activity/activity_canonicalHRF_allSubjs_replication.csv',delimiter=',')
# load fc data
connectivity = np.loadtxt(resultdir + '/connectivity/conn_taskVrest_gbc_replication_FIR.csv', delimiter=',')
# load timescale data
timescales = np.loadtxt(resultdir + '/timescales/Murray_taus_regions_autocorrelation_rest100_replication.txt',delimiter=',')
bad_ind = np.where(timescales>50)
timescales[bad_ind] = np.nan

nsubjs = subj_activity.shape[1]
```

#### Load myelin

```python
myelinmap = np.squeeze(nib.load(resultdir + 'Mean.339.MyelinMap_BC_MSMAll.32k_fs_LR.dscalar.nii').get_data())
nvertices = int(glasser2.shape[0]/2)

glasser_lr = np.hstack((glasser2[nvertices:],glasser2[:nvertices])) # need to flip this original parcellation

myelinmap_parc = np.zeros((nParcels,))
for roi in range(nParcels):
    roi_ind = np.where(glasser_lr==roi+1)[0]
    myelinmap_parc[roi] = np.mean(myelinmap[roi_ind])
```

# Correlation with task activations and myelin content

```python
r, lower_ci, upper_ci = bootstrap_corr(t_act,myelinmap_parc,corr_type='spearman')
print('Spearman True:', r, '| CI = (', lower_ci, ',', upper_ci,')')
#
r, lower_ci, upper_ci = bootstrap_corr(t_act,myelinmap_parc,corr_type='pearson')
print('Pearson True:', r, '| CI = (', lower_ci, ',', upper_ci,')')

plt.figure(figsize=(10,7))
sns.regplot(t_act,myelinmap_parc,color='k',fit_reg=False)
sns.regplot(t_act,myelinmap_parc,color='k',scatter_kws={'color':roiColorsByNetwork})
plt.ylabel('Myelin content (T1w/T2w)',fontsize=28)
plt.xlabel('Task activation (abs. t-value)',fontsize=28)
plt.title('Myelin content vs.\n task activation change',fontsize=32)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.tight_layout()
# plt.savefig('../figures/manuscriptNetDim2b_Final/Fig4s_MyelinXActivity.pdf')
```

```python
r, lower_ci, upper_ci = bootstrap_corr(myelinmap_parc,np.mean(connectivity,axis=1),corr_type='spearman')
print('Spearman True:', r, '| CI = (', lower_ci, ',', upper_ci,')')
#
r, lower_ci, upper_ci = bootstrap_corr(myelinmap_parc,np.mean(connectivity,axis=1),corr_type='pearson')
print('Pearson True:', r, '| CI = (', lower_ci, ',', upper_ci,')')

plt.figure(figsize=(10,7))
sns.regplot(np.mean(connectivity,axis=1),myelinmap_parc,color='k',fit_reg=False)
sns.regplot(np.mean(connectivity,axis=1),myelinmap_parc,color='k',scatter_kws={'color':roiColorsByNetwork})
plt.ylabel('Myelin content (T1w/T2w)',fontsize=28)
plt.xlabel('FC change (r)',fontsize=28)
plt.title('Myelin content vs.\n task FC change',fontsize=32)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.tight_layout()
# plt.savefig('Fig4s_MyelinXConnectivity.pdf')
```

```python
r, lower_ci, upper_ci = bootstrap_corr(myelinmap_parc,np.nanmean(timescales,axis=1),corr_type='spearman')
print('Spearman True:', r, '| CI = (', lower_ci, ',', upper_ci,')')
#
r, lower_ci, upper_ci = bootstrap_corr(myelinmap_parc,np.nanmean(timescales,axis=1),corr_type='pearson')
print('Pearson True:', r, '| CI = (', lower_ci, ',', upper_ci,')')

plt.figure(figsize=(10,7))
sns.regplot(np.nanmean(timescales,axis=1),myelinmap_parc,color='k',fit_reg=False)
sns.regplot(np.nanmean(timescales,axis=1),myelinmap_parc,color='k',scatter_kws={'color':roiColorsByNetwork})
plt.ylabel('Myelin content (T1w/T2w)',fontsize=28)
plt.xlabel(r'Intrinsic timescale ($\tau$)',fontsize=28)
plt.title('Myelin content vs. intrinsic timescale',fontsize=32)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.tight_layout()
# plt.savefig('Fig4s_MyelinXTimescale.pdf')
```
